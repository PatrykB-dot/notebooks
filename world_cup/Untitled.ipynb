```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv('data/WorldCupMatches.csv')
```

Drop na values from dataframe

```python
df.dropna(how='all', inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

Top 20 Home Team Names

```python
home_teams = df['Home Team Name'].value_counts()
```

```python
home_teams[:20].plot.bar(figsize=(20, 5))
```

Top 20 Away Team Names

```python
away_teams = df['Away Team Name'].value_counts()
```

```python
away_teams[:20].plot.bar(figsize=(20, 5))
```

Which game had most attendance ?

```python
# Search index of max value in attendance column
highest_attendance_index = df['Attendance'].idxmax()
```
