```python
import pandas as pd
import numpy as np
```

```python
sf_permits = pd.read_csv("data/Building_Permits.csv")
```

```python
# take a look to data, get 5 random columns 
sf_permits.sample(5)
```

Check missing points

```python
# get the number of 
missing_values_count = sf_permits.isnull().sum()
```

I want to see how many missing values i have in dataframe

```python
total_cells = np.prod(sf_permits.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
(total_missing/total_cells) * 100
```

Almost 26 percent of data has null values


I want to use dropna function and compare data sets with null and without null values

```python
columns_with_na_dropped = sf_permits.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# just how much data did i lose?
print("Columns in original dataset: %d \n" % sf_permits.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```

Looks like i lost many columns by dropping na values by this method

Filling missing values

```python
subset_sf_permits = sf_permits.loc[:, 'Existing Construction Type':'Record ID'].head()
subset_sf_permits
```

```python
subset_sf_permits.fillna(method = 'bfill', axis=0).fillna(0)
```
